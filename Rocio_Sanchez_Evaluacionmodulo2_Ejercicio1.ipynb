```python

url = "https://beta.adalab.es/resources/apis/pelis/pelis.json"

datos = requests.get(url)
peliculas_json = datos.json()   

print(datos.status_code)
print(datos.content)

```

```python
df = pd.DataFrame(peliculas_json)
```
